```python
#데이터 시각화
import seaborn as sns
import matplotlib.pyplot as plt



#1번째 공격 유형 그래프 전체적으로 보기(nomal포함해서)
# 도화지를 아주 넓게 설정
plt.figure(figsize=(40, 15))

order_list = df['attack'].value_counts().index
plt.subplot(2, 2, 1)
ax = sns.countplot(data=df, x='attack', order=order_list, palette='viridis')
plt.title("전체 공격 유형")

plt.xticks(rotation=45)


plt.tight_layout()
plt.show()

#2번째 전체 프로토콜 타입 비율 확인하기

#2번째 Dos 공격 유형에서 비율을 파이차트로 보기
#너무 공격 유형이 너무 많다.... 칼럼도 너무 많아서 공격 유형이 너무 자잘하게 나뉘어 있어서 머신러닝 학습에 방해가 될 수 있겠다.
#전처리가 필요하다 !! 중요한 칼럼만 가져오고, 중요하지 않은 칼럼은 빼고 18가지
#공격 유형 21가지를 nomal, Dos, Probe, R2L, U2R로 분류
```

```python
# 1.
selected_features = [
    # 기본 정보
    'duration', 'protocol_type', 'service', 'flag', 'src_bytes', 'dst_bytes',
    # 콘텐츠/권한 (R2L, U2R과 관련된 칼럼)
    'logged_in', 'wrong_fragment', 'root_shell', 'hot', 'num_failed_logins',
    # 트래픽 패턴 (DoS, Probe과 관련된 칼럼)
    'count', 'serror_rate', 'same_srv_rate', 'diff_srv_rate',
    # 호스트 기록
    'dst_host_srv_count', 'dst_host_same_src_port_rate', 'dst_host_serror_rate'
]

#18개의 중요한 칼럼이 있는 df생성
new_df = df[selected_features + ['attack']].copy()

#attack 5가지로 치환
attack_mapping = {
    # Normal
    'normal': 'Normal',

    # DoS (서비스 거부 공격)
    'neptune': 'DoS', 'back': 'DoS', 'land': 'DoS', 'pod': 'DoS',
    'smurf': 'DoS', 'teardrop': 'DoS',

    # Probe (정보 수집 및 정찰)
    'ipsweep': 'Probe', 'nmap': 'Probe', 'portsweep': 'Probe', 'satan': 'Probe',

    # R2L (원격지에서 로컬 권한 획득 시도 - 주로 로그인 관련)
    'ftp_write': 'R2L', 'guess_passwd': 'R2L', 'imap': 'R2L', 'multihop': 'R2L',
    'phf': 'R2L', 'spy': 'R2L', 'warezclient': 'R2L', 'warezmaster': 'R2L',

    # U2R (일반 유저 권한으로 루트 권한 획득 시도 - 가장 위험도가 높은 공격)
    'buffer_overflow': 'U2R', 'loadmodule': 'U2R', 'perl': 'U2R', 'rootkit': 'U2R'
}


new_df['attack'] = new_df['attack'].map(attack_mapping)

print("\n데이터 미리보기:")


#데이터프레임 내용 확인
print(new_df.info())

#데이터 통계 정보 확인
print(new_df.describe())

#각 열이 가지고 있는 원소개수
print(new_df.count())
print("\n데이터 미리보기:")
new_df.head()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 시각화 스타일 설정 (가독성을 높임)
sns.set_theme(style="whitegrid")

# 그래프를 그릴 캔버스 크기 설정 (가로 14, 세로 6)
plt.figure(figsize=(20, 6))

# -------------------------------------------------------
# 1. 공격 유형별 데이터 분포 (왼쪽 그래프)
# -------------------------------------------------------
plt.subplot(1, 2, 1) # 1행 2열 중 첫 번째
ax1 = sns.countplot(x='attack', data=new_df, palette='viridis', order=['Normal', 'DoS', 'Probe', 'R2L', 'U2R'])
plt.title('Distribution of Attack Types', fontsize=15)
plt.xlabel('Attack Type')
plt.ylabel('Count')

# 막대 위에 숫자 표시 (데이터 개수를 정확히 보기 위해)
for p in ax1.patches:
    ax1.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha = 'center', va = 'center', xytext = (0, 9), textcoords = 'offset points')

# -------------------------------------------------------
# 2. 공격 유형별 로그인 여부 비교 (오른쪽 그래프)
# -------------------------------------------------------
plt.subplot(1, 2, 2) # 1행 2열 중 두 번째
# hue='logged_in' 옵션을 주면 로그인 여부(0, 1)로 막대를 나누어 그립니다.
sns.countplot(x='attack', hue='logged_in', data=new_df, palette='Set2', order=['Normal', 'DoS', 'Probe', 'R2L', 'U2R'])
plt.title('Logged In Status by Attack Type', fontsize=15)
plt.xlabel('Attack Type')
plt.ylabel('Count')
plt.legend(title='Logged In', labels=['No (0)', 'Yes (1)'])

# 레이아웃 조정 및 출력
plt.tight_layout()
plt.show()
```
